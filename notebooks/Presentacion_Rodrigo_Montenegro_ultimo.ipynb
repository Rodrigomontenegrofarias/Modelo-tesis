```python
import numpy as np
import numpy
import matplotlib.image as img
import matplotlib.pyplot as plot
import cv2 
import PIL 
import matplotlib.pyplot as plt
from skimage import color
from skimage import io 

data = {
    'Frame': [],
    'Rojo Vino': [],
    'Rojo Intenso': [],
    'Cafe Oscuro': [],
    'Cafe': [],
    'Mostaza': [],
    'Cafe Piel': [],
    'Amarillo': [],
    'Verde Claro': [],
    'Verde Oscuro': [],
    'Celeste': [],
    'Azul': [],
    'Azul Marino': [],
    'Azul Oscuro': [],
    'Rosado': [],
    'Morado': []
}


# Llamar a la función pasando todas las máscaras como argumentos

for i in range(imageGif6.n_frames):
    #Muestra el numero en funicon de  la iteracion.
    
    colores_no_encontrados = []
    # Cargar la imagen
    image = cv2.imread("frame"+str(i)+".png")
    
    # Eliminar los 4 píxeles de la esquina de las imagenes.
    image[0:1, 0:1] = (0, 0, 0)
    image[0:1, -1:] = (0, 0, 0)
    image[-1:, 0:1] = (0, 0, 0)
    image[-1:, -1:] = (0, 0, 0)
    
    # Convertir la imagen al formato HSV
    image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    plt.imshow(cv2.cvtColor(image_hsv, cv2.COLOR_HSV2RGB))
    plt.show()
    #####muestra lospixeles.
    

    #Empezamos con la clasificacion de colores en formatos hsv.                                                                                                                                                                       

    ####################################################
    # Definir los rangos de color para la segmentación #
    ####################################################

    #####################
    #   ROJO Intenso 100%  #
    #####################    
    #rojo 1
    rojo_intenso_bajo1 = np.array([0, 100, 217], np.uint8)
    rojo_intenso_alto1 = np.array([8, 255, 255], np.uint8)
    # rojo 2
    rojo_intenso_bajo2 = np.array([170, 100, 217], np.uint8)
    rojo_intenso_alto2 = np.array([179, 255, 255], np.uint8)
    #####################
    #   ROJO Vino 94%   #
    #####################
    #rojo vino 1
    rojo_vino_bajo = np.array([170, 100, 20], np.uint8)
    rojo_vino_alto = np.array([179, 255, 216], np.uint8)
    #rojo vino 2
    rojo_vino_bajo1 = np.array([0, 100, 20], np.uint8)
    rojo_vino_alto1 = np.array([8, 255, 216], np.uint8)
    ######################
    #   CAFE oscuro 87%  #
    ######################
    cafe_oscuro_bajo = np.array([10, 240, 78], np.uint8)
    cafe_oscuro_alto = np.array([14, 255, 175], np.uint8) 
    ###############
    #   CAFE 81%  #
    ###############
    cafe_bajo = np.array([18, 234, 119], np.uint8)
    cafe_alto = np.array([23, 255, 133], np.uint8)
    #######################
    #   CAFE Mostasa 75%  #
    #######################  
    cafe_mostaza_bajo = np.array([17, 155, 183], np.uint8)
    cafe_mostaza_alto = np.array([22, 234, 250], np.uint8)
    ####################
    #   CAFE Piel 68%  #
    ####################  
    cafe_bajo_piel = np.array([1, 89, 204], np.uint8)
    cafe_alto_piel = np.array([16, 233, 255], np.uint8)
    ####################
    #   Amarillo 62%   #
    ####################
    amarillo_claro_bajo = np.array([23, 97, 227], np.uint8)
    amarillo_claro_alto = np.array([31, 255, 255], np.uint8)   
    ########################
    #   Verde Claro  56%   #
    ######################## 
    verde_claro_bajo = np.array([55, 203, 118], np.uint8)
    verde_claro_alto = np.array([73, 255, 215], np.uint8)
    ##################
    #   Verde  48%   #  
    ##################    
    verde_bajo = np.array([25, 95, 92], np.uint8)
    verde_alto = np.array([70, 255, 115], np.uint8)
    #########################
    #   Verde Oscuro  43%   #
    #########################
    verde_oscuro_bajo = np.array([28, 50, 50], np.uint8)
    verde_oscuro_alto = np.array([133, 226, 88], np.uint8)
    #########################
    #   Celeste  %37        #
    #########################
    celeste_oscuro_bajo = np.array([97, 166, 184], np.uint8)
    celeste_claro_alto = np.array([113, 253, 253], np.uint8)                                      
    #########################
    #   Azul     %30        #
    #########################   [125 255 203# 119
    azul_bajo = np.array([119, 211, 127], np.uint8)
    azul_alto = np.array([124, 255, 230], np.uint8)
    ######################### 
    #   Azul Marino  %24    #
    ######################### 
    azul_marino_bajo = np.array([118, 247, 91], np.uint8)
    azule_marino_alto = np.array([124, 255, 112], np.uint8)
    #########################
    #   Azul Oscuro  %18    #
    ######################### 
    azul_oscuro_bajo = np.array([106, 245, 49], np.uint8)
    azul_oscuro_alto = np.array([125, 255, 94], np.uint8)
    #########################
    #
    #
    #########################
    Rosado_oscuro_bajo = np.array([151, 159, 50], np.uint8)
    Rosado_oscuro_alto = np.array([158, 254, 193], np.uint8)
    #########################  [162 234  98]
    #   Morado     %5       #
    #########################    
    #Reacer dos
    Morado_oscuro_bajo = np.array([125, 68, 50], np.uint8)#120 -68
    Morado_oscuro_alto = np.array([150, 255, 203], np.uint8)
    #
    Morado_oscuro_bajo1 = np.array([159, 159, 18], np.uint8)
    Morado_oscuro_alto1 = np.array([176, 255, 187], np.uint8)
    #######
    ##############################################
    # Aplicar la segmentación de color o máscara #  
    ##############################################
    
    #ROJO Intenso 100%#
    mascara_rojo_1 = cv2.inRange(image_hsv, rojo_vino_bajo, rojo_vino_alto)
    mascara_rojo_3 = cv2.inRange(image_hsv, rojo_vino_bajo1, rojo_vino_alto1)
    #ROJO Vino 94%#
    mascara_rojo_2 = cv2.inRange(image_hsv, rojo_intenso_bajo1, rojo_intenso_alto1)
    mascara_rojo_4 = cv2.inRange(image_hsv, rojo_intenso_bajo2, rojo_intenso_alto2)
    #CAFE 87%#  
    mascara_cafe_oscuro = cv2.inRange(image_hsv, cafe_oscuro_bajo, cafe_oscuro_alto)
    #CAFE 81%#
    mascara_cafe = cv2.inRange(image_hsv, cafe_bajo, cafe_alto)
    #CAFE Mostasa 75%#    
    mascara_cafe_mostaza = cv2.inRange(image_hsv, cafe_mostaza_bajo, cafe_mostaza_alto)
    #   CAFE Piel 68%  #  
    mascara_cafe_piel = cv2.inRange(image_hsv, cafe_bajo_piel, cafe_alto_piel)
    #   Amarillo 62%   # 
    mascara_Amarillo = cv2.inRange(image_hsv, amarillo_claro_bajo, amarillo_claro_alto)
    #   Verde Claro  56%   #
    mascara_VerdeClaro = cv2.inRange(image_hsv, verde_claro_bajo, verde_claro_alto)
    #   Verde  48%   #     
    mascara_verde = cv2.inRange(image_hsv, verde_bajo, verde_alto)
    #   Verde Oscuro  43%   #
    mascara_VerdeOscuro = cv2.inRange(image_hsv, verde_oscuro_bajo, verde_oscuro_alto)
    #   Celeste  %37        #
    mascara_Celeste = cv2.inRange(image_hsv, celeste_oscuro_bajo, celeste_claro_alto)
    #   Azul     %30        #
    mascara_azul = cv2.inRange(image_hsv, azul_bajo, azul_alto)
    #   Azul Marino  %24    #
    mascara_azulMarino = cv2.inRange(image_hsv, azul_marino_bajo, azule_marino_alto)
    #   Azul Oscuro  %18    #
    mascara_AzulOscuro = cv2.inRange(image_hsv, azul_oscuro_bajo, azul_oscuro_alto)
    #   Rosado    %11       #
    mascara_Rosado = cv2.inRange(image_hsv, Rosado_oscuro_bajo, Rosado_oscuro_alto)
    #   Morado     %5       #
    mascara_morado = cv2.inRange(image_hsv, Morado_oscuro_bajo, Morado_oscuro_alto)
    
    mascara_morado1 = cv2.inRange(image_hsv, Morado_oscuro_bajo1, Morado_oscuro_alto1)
    

    ###############################################################################
    # Calcular el número total de píxeles después de eliminar los de las esquinas #
    ###############################################################################     

    total_pixeles = (image_hsv.size - 12) // 3  # Dividido por 3 para considerar canales RGB

    ######################################
    # Calcular los píxeles de cada color #
    ######################################
    
    #   ROJO Intenso 100%  #
    pixeles_rojo_1 = cv2.countNonZero(mascara_rojo_1)
    pixeles_rojo_3 = cv2.countNonZero(mascara_rojo_3)
    #   ROJO Vino 94%   #
    pixeles_rojo_2 = cv2.countNonZero(mascara_rojo_2)
    pixeles_rojo_4 = cv2.countNonZero(mascara_rojo_4)
    #   CAFE 87%  #
    pixeles_cafe_oscuro = cv2.countNonZero(mascara_cafe_oscuro)
    #   CAFE 81%  #  
    pixeles_cafe = cv2.countNonZero(mascara_cafe)
    #   CAFE Mostasa 75%  #
    pixeles_cafe_mostaza = cv2.countNonZero(mascara_cafe_mostaza)
    #   CAFE Piel 68%  #
    pixeles_cafe_piel = cv2.countNonZero(mascara_cafe_piel)
    #   Amarillo 62%   #
    pixeles_amarillo = cv2.countNonZero(mascara_Amarillo)
    #   Verde Claro  56%   #
    pixeles_verdeclaro = cv2.countNonZero(mascara_VerdeClaro)
    #   Verde  48%   #    
    pixeles_verde = cv2.countNonZero(mascara_verde)
    #   Verde Oscuro  43%   # 
    pixeles_verdeoscuro = cv2.countNonZero(mascara_VerdeOscuro)
    #   Celeste  %37        #
    pixeles_Celeste = cv2.countNonZero(mascara_Celeste)
    #   Azul     %30        #
    pixeles_azul = cv2.countNonZero(mascara_azul)
    #   Azul Marino  %24    #
    pixeles_azulMarino = cv2.countNonZero(mascara_azulMarino)
    #   Azul Oscuro  %18    #
    pixeles_azulOscuro = cv2.countNonZero(mascara_AzulOscuro)
    #   Rosado    %11       #
    pixeles_Rosado = cv2.countNonZero(mascara_Rosado)
    #   Morado     %5       #
    pixeles_Morado = cv2.countNonZero(mascara_morado)
    pixeles_Morado1 = cv2.countNonZero(mascara_morado1)
   
    #########################################################
    ## Calcular los porcentajes de píxeles para cada color###
    #########################################################

    #   ROJO Intenso 100%  #
    porcentaje_rojo_vino = (pixeles_rojo_1 / total_pixeles) * 100
    porcentaje_rojo_intenso = (pixeles_rojo_2 / total_pixeles) * 100
    #   ROJO Vino 94%   #
    porcentaje_rojo_vino1 = (pixeles_rojo_3 / total_pixeles) * 100
    porcentaje_rojo_intenso1 = (pixeles_rojo_4 / total_pixeles) * 100
    #   CAFE 87%  #
    porcentaje_cafe_oscuro = ( pixeles_cafe_oscuro / total_pixeles) * 100
    #   CAFE 81%  #
    porcentaje_cafe = ( pixeles_cafe / total_pixeles) * 100
    #   CAFE Mostasa 75%  #
    porcentaje_cafe_mostaza = (  pixeles_cafe_mostaza / total_pixeles) * 100
    #   CAFE Piel 68%  #
    porcentaje_cafe_piel = (  pixeles_cafe_piel / total_pixeles) * 100
    #   Amarillo 62%   #
    porcentaje_amarillo = (  pixeles_amarillo / total_pixeles) * 100
    #   Verde Claro  56%   #
    porcentaje_verdeclaro = (  pixeles_verdeclaro / total_pixeles) * 100 
    #   Verde  48%   #   
    porcentaje_verde = (  pixeles_verde / total_pixeles) * 100
    #   Verde Oscuro  43%   #
    porcentaje_verdeoscuro = (  pixeles_verdeoscuro / total_pixeles) * 100
    #   Celeste  %37        #
    porcentaje_celeste = (  pixeles_Celeste / total_pixeles) * 100
    #   Azul     %30        #
    porcentaje_azul = (  pixeles_azul / total_pixeles) * 100
    #   Azul Marino  %24    #
    porcentaje_azulmarino = ( pixeles_azulMarino  / total_pixeles) * 100
    #   Azul Oscuro  %18    #
    porcentaje_azuloscuro = (  pixeles_azulOscuro / total_pixeles) * 100
    #   Rosado    %11       #
    porcentaje_rosado = ( pixeles_Rosado  / total_pixeles) * 100
    #   Morado     %5       #
    porcentaje_morado = ( pixeles_Morado  / total_pixeles) * 100
    porcentaje_morado1 = ( pixeles_Morado1  / total_pixeles) * 100
    ## Calcular los porcentajes de píxeles para cada color
    

    #   ROJO Intenso 100%  #
    porcentaje_rojo_intenso_total = porcentaje_rojo_intenso + porcentaje_rojo_intenso1
    pixeles_rojo_intenso_total = pixeles_rojo_2 + pixeles_rojo_4
    #   ROJO Vino 94%   #
    porcentaje_rojo_vino_total = porcentaje_rojo_vino + porcentaje_rojo_vino1
    pixeles_rojo_vino_total = pixeles_rojo_1 + pixeles_rojo_3
    #   CAFE 87%            #
    #   CAFE 81%            #
    #   CAFE Mostasa 75%    #
    #   CAFE Piel 68%       #
    #   Amarillo 62%        #
    #   Verde Claro  56%    #
    #   Verde  48%          #
    #   Verde Oscuro  43%   #
    #   Celeste  %37        #
    #   Azul     %30        #
    #   Azul Marino  %24    #
    #   Azul Oscuro  %18    #
    #   Rosado    %11       #
    #   Morado     %5       #
    porcentaje_morado_total = porcentaje_morado + porcentaje_morado1
    pixeles_morado_total = pixeles_Morado + pixeles_Morado1
    #########################
    

    # suma de pixel por frame
    SUMA_PORCENTAJE_PIXEL= pixeles_rojo_intenso_total + pixeles_rojo_vino_total + pixeles_morado_total + pixeles_cafe_oscuro + pixeles_cafe + pixeles_cafe_mostaza + pixeles_cafe_piel + pixeles_amarillo +  pixeles_verdeclaro + pixeles_Celeste + pixeles_verdeoscuro + pixeles_azul + pixeles_azulMarino + pixeles_azulOscuro + pixeles_Rosado + pixeles_verde

    SUMA_PORCENTAJE_total= porcentaje_rojo_intenso_total + porcentaje_rojo_vino_total + porcentaje_morado_total + porcentaje_cafe_oscuro + porcentaje_cafe + porcentaje_cafe_mostaza + porcentaje_cafe_piel + porcentaje_amarillo  + porcentaje_verdeclaro + porcentaje_verde + porcentaje_verdeoscuro + porcentaje_celeste + porcentaje_azul + porcentaje_azulmarino + porcentaje_azuloscuro + porcentaje_rosado 
    

    print(f"Frame {i}:")
    
    if SUMA_PORCENTAJE_PIXEL == 32 :
       
        print("Porcentaje de píxeles y porcentaje total:",SUMA_PORCENTAJE_total,SUMA_PORCENTAJE_PIXEL)  
        print("------")
    else:
        print("------")
        
        ## PRINT EN PANTALLA.
        print(image_hsv)
        print("---RESUMEN---")
        print("Porcentaje de píxeles rojo vino:", porcentaje_rojo_vino_total,pixeles_rojo_vino_total)
        print("Porcentaje de píxeles rojo intenso:",porcentaje_rojo_intenso_total,pixeles_rojo_intenso_total)
        print("Porcentaje de píxeles cafe oscuro:",porcentaje_cafe_oscuro,pixeles_cafe_oscuro)
        print("Porcentaje de píxeles cafe:",porcentaje_cafe,pixeles_cafe)
        print("Porcentaje de píxeles mostaza:",porcentaje_cafe_mostaza,pixeles_cafe_mostaza)
        print("Porcentaje de píxeles piel:",porcentaje_cafe_piel,pixeles_cafe_piel)
        print("Porcentaje de píxeles Amarillo:",porcentaje_amarillo,pixeles_amarillo)
        print("Porcentaje de píxeles Verde Claro:",porcentaje_verdeclaro,pixeles_verdeclaro)
        print("Porcentaje de píxeles Verde:",porcentaje_verde,pixeles_verde)    
        print("Porcentaje de píxeles Verde oscuro:",porcentaje_verdeoscuro,pixeles_verdeoscuro)
        print("Porcentaje de píxeles Celeste:",porcentaje_celeste,pixeles_Celeste)
        print("Porcentaje de píxeles Azul :",porcentaje_azul,pixeles_azul)
        print("Porcentaje de píxeles Azul Marino :",porcentaje_azulmarino,pixeles_azulMarino)
        print("Porcentaje de píxeles Azul Oscuro :", porcentaje_azuloscuro,pixeles_azulOscuro)
        print("Porcentaje de píxeles Rosado:",porcentaje_rosado,pixeles_Rosado)
        print("Porcentaje de píxeles Morado:",porcentaje_morado_total,pixeles_morado_total)
        print("------")
        print("Porcentaje de píxeles y porcentaje total:",SUMA_PORCENTAJE_total,SUMA_PORCENTAJE_PIXEL)  
        print("------")
    
    # Guardar los datos en el diccionario
    data['Frame'].append(i)
    data['Rojo Vino'].append(pixeles_rojo_vino_total * 95)
    data['Rojo Intenso'].append(pixeles_rojo_intenso_total * 100)
    data['Cafe Oscuro'].append(pixeles_cafe_oscuro * 87)
    data['Cafe'].append(pixeles_cafe * 81)
    data['Mostaza'].append(pixeles_cafe_mostaza * 75)
    data['Cafe Piel'].append(pixeles_cafe_piel * 62)
    data['Amarillo'].append(pixeles_amarillo * 56)
    data['Verde Claro'].append(pixeles_verdeclaro * 48)
    data['Verde Oscuro'].append(pixeles_verdeoscuro * 43)
    data['Celeste'].append(pixeles_Celeste * 37)
    data['Azul'].append(pixeles_azul * 30)
    data['Azul Marino'].append(pixeles_azulMarino * 24)
    data['Azul Oscuro'].append(pixeles_azulOscuro * 18)
    data['Rosado'].append(pixeles_Rosado * 11)
    data['Morado'].append(pixeles_morado_total * 5)
    
# Crear un DataFrame a partir del diccionario de imagenes

df = pd.DataFrame(data)
columns_to_sum = ['Rojo Vino', 'Rojo Intenso', 'Cafe Oscuro', 'Cafe', 'Mostaza', 'Cafe Piel', 'Amarillo', 'Verde Claro', 'Verde Oscuro', 'Celeste', 'Azul', 'Azul Marino', 'Azul Oscuro', 'Rosado', 'Morado']
df['Nubosidad-frame'] = df[columns_to_sum].sum(axis=1) / 32

# Mostrar el DataFrame
print(df)
```
